```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = '/content/drive/MyDrive/kc_house_data.csv'
df = pd.read_csv(df)
df.info()
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns= ['id','price','date','lat','long'])
```

```python
X.shape
```

```python
y = df['price']
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 3)
```

```python
clf = RandomForestRegressor(n_estimators= 100, bootstrap= True, oob_score= True)
```

```python
clf
```

```python
clf.fit(X_train, y_train)
```

```python
score_random_forest_train = clf.score(X_train,y_train)
score_random_forest_train
```

```python
score_random_forest_test = clf.score(X_test, y_test)
score_random_forest_test
```

```python
clf.oob_score_
```

```python
clf.feature_importances_
```

```python
clf_features = pd.DataFrame({'feature': X_train.columns,'feature_importance': clf.feature_importances_})
clf_features.sort_values(by = 'feature_importance', ascending= False)
```

Bagging model

```python
bagg_tree = BaggingRegressor()
bagg_tree.fit(X_train, y_train)
```

```python
bagg_tree_train_score = bagg_tree.score(X_train, y_train)
bagg_tree_train_score
```

```python
bagg_tree_test_score = bagg_tree.score(X_test, y_test)
bagg_tree_test_score
```

```python
price_prediction = pd.DataFrame({'actual price': y_test,'price_predition': bagg_tree.predict(X_test)})
price_prediction
```
